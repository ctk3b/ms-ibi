```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

target_rdf = np.loadtxt('../rdfs/lmp.target.txt')

ax1.plot(target_rdf[:,0], target_rdf[:,1], 'k-', label='target')

for step in range(10):
    try:
        rdf = np.loadtxt('../rdfs/pair_1-1-state_state0-step{}.txt'.format(step))
        pot = np.loadtxt('../potentials/step{}.pot.1-1.txt'.format(step), skiprows=3)
        ax1.plot(rdf[:,0], rdf[:,1], label=step)
        ax2.plot(pot[:,1], pot[:,2], '--', label=step)
        print('Found step {} ...'.format(step))
    except FileNotFoundError:
        print('Did not find step {} ...'.format(step))

ax1.legend(loc=0)
ax2.legend(loc=0)
ax2.set_ylim((-2, 1))
ax1.set_ylabel('RDF')
ax2.set_ylabel('V(r)')
ax1.set_xlabel('r')
ax2.set_xlabel('r')
plt.show()
```
